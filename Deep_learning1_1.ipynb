  #                              1. A SIGMOID FUNCTION
  *Running Sigmoid Function for X (inputs) and W (Weight)*

```python
#importing numpy for matric operation
import numpy as np
#importing tensorflow to use sigmoid function
import tensorflow as tf
#importing math 
import math


X = np.array([[0,1,-7],[8,-9,56],[90,-89,67]]); W = np.array([[0,1,7],[8,9,56],[90,89,67]]);
z = tf.constant(np.dot(X,W.transpose()), dtype = tf.float32)
print(z)

y = tf.math.sigmoid(z, name="sigmoid")

print(y)

#To plot sigmoid function
#importing required module
import matplotlib.pyplot as plt
#y vs z

plt.plot(z,y,)

#naming axis
plt.xlabel('z - axis')
plt.ylabel('y = g(z)')
#title
plt.title("SIGMOID FUNCTION")
plt.show
```

1.B **Graph of Actual Sigmoid Function**

```python
# importing the required module
import matplotlib.pyplot as plt
import math
import tensorflow as tf

y= tf.math.sigmoid(z,name="sigmoid")

z = np.linspace(-10,10,100)
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
# plotting the points
plt.plot(z, y,'')

# naming the x axis
plt.xlabel('z - axis')
# naming the y axis
plt.ylabel('y = g(z)')
plt.title("1. SIGMOID Function")
```

#2. Hyperbolic Tangent

```python
#importing numpy for matric operation
import numpy as np
#importing tensorflow to use sigmoid function
import tensorflow as tf
#importing math 
import math


X = np.array([[0,1,-7],[8,-9,56],[90,-89,67]]); W = np.array([[0,1,7],[8,9,56],[90,89,67]]);
z = tf.constant(np.dot(X,W.transpose()), dtype = tf.float32)
print(z)

y = tf.math.tanh(z,name="Hyperbolic")

print(y)

#To plot sigmoid function
#importing required module
import matplotlib.pyplot as plt
#y vs z

plt.plot(z,y,)

#naming axis
plt.xlabel('z - axis')
plt.ylabel('y = g(z)')
#title
plt.title("HYPERBOLIC FUNCTION")
plt.show
```

```python
# importing the required module
import matplotlib.pyplot as plt
import math
import tensorflow as tf

y= tf.math.tanh(z,name="HYPERBOLIC")

z = np.linspace(-10,10,100)
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
# plotting the points
plt.plot(z, y,)

# naming the x axis
plt.xlabel('z - axis')
# naming the y axis
plt.ylabel('y = g(z)')
plt.title("2. HYPERBOLIC FUNCTION")
```

#3. Rectified Linear Unit (ReLU)

```python
# importing the required module
import matplotlib.pyplot as plt
import math
import tensorflow as tf

y = tf.nn.relu(z, name= "ReLU")

z = np.linspace(-10,10,100)
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
# plotting the points
plt.plot(z, y,)

# naming the x axis
plt.xlabel('z - axis')
# naming the y axis
plt.ylabel('y = g(z)')
plt.title("3. Rectified Linear Unit (ReLU)")
```
